## What about maps and spatial data?

Let's import the treaties map

```python
import geopandas as gpd 
```

```python
treaties_map = gpd.read_file('treaties_map.geojson')
treaties_map.head()
```

```python
treaties_map.plot('color')
```

```python
treaties_map.explore()
```

```python
treaties_joined = treaties_map.merge(treaties, left_on='id', right_on='id')
```

```python
treaties_joined.head()
```

```python
treaties_joined.explore('color_x')
```
